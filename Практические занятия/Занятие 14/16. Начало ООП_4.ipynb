# Задачи на ООП

# Задача 16 $\star$

**Написать собственный класс ```Scan``` (```Скан```).**

**В качестве необходимых атрибутов класс ```Scan``` должен имплементировать:**

* Уникальное имя скана ```name```;
* Атрибут ```points``` содержащий коллекцию объектов ```ScanPoint```;
* Атрибут ```borders``` содержащий в себе данные о границах скана ($x_{min}, x_{max}, y_{min}, y_{max}, \dots$);
* Атрибут ```parser``` содержащий объект, реализующий в себе логику чтения исходного файла с данными и заполнения полей ```points```, ```borders```, и тп. Например класс ```ScanParserFromTXT``` "читающий" скан из текстового файла, или ```ScanParserFromLAS``` "читающий" скан из файла скана с расширением ```.las```. Логика работы классов ```ScanParser***``` должна реализовывать внутренний метод ```parse```, возвращающий в содержащий его объект скана необходимые данные, и вызываемый в объекте ```Scan``` при заполнении скана точками;
* Атрибут ```plotter``` отвечающий за логику вывода точек скана на график (объект должен иметь возможность строить одновременно только часть точек из скана);

**Переопределите стандартный метод приведения объекта к строке ```__str__(self)``` и ```__repr__(self)``` (методы должны вернуть объект строки ```str```, описывающий объект в котором он вызван).**

**Переопределите стандартный метод ```__len__(self)``` который определяет логику работы стандартной функции ```len()```. Метод должен возвращать количество точек, содержащихся в скане.**

**Переопределите стандартный метод ```__iter(self)___``` который определяет логику работы стандартной функции ```iter()```. Метод должен возвращать объект итератора для последовательного перебора коллекции ```points```.**

**Напишите метод ```point_filter(self, pf, replace=True)```, где:**
* ```pf``` - объект определяющий логику фильтрации или разрежения исходного скана (например ```ScanDecimator``` - объект который будет разрежать скан, исходя из заданного процента точек, который необходимо оставить в скане или ```ScanPolygonFilter``` - который оставит в скане только те точки, которые находятся внутри заданной полилинии и тп);
* ```replace``` - определяет, заменяются ли данные в исходном объекте (если ```True```) или (если ```False```) метод возвращает новый объект класса ```Scan``` (имя скана ```name``` можно сгенерировать через агрегацию имени исходного скана и суффикс, определяемый способом фильтрации (например ```scan_1``` в котором через ```ScanDecimator``` осталось $10\%$ точек можно назвать ```scan_1_decim_10``` и тп) саму логику возвращения нового скана можно реализовать через изменение копии исходного скана (через метод ```copy()``` или ```deepcopy()```.

> Постарайтесь придумать и реализовать максимальное количество способов фильтрации сканов для разных условий. (Например оставляющие только точки в определенном радиусе от заданной точки или оставляющие только точки определенного цвета (не конкретного значения оттенка, а именно цвета), и тп)
