# Задачи на базовое ООП

## Задача 13.1 $\star$

**Написать собственный класс ```Point``` (```Точка```).**

В качестве необходимых атрибутов класс ```Point``` должен имплементировать:

* Уникальный для каждой точки идентификатор ```id```, определяющий порядковый номер создаваемой точки (Можно реализовать через инкременитруемый счетчик в *статической переменной ```count```* реализованный на уровне класса);
* Координаты ```x```, ```y```, ```z``` (для плоских координат ```z``` можно принимать по умолчанию равной ```0```);
* Имя точки ```name``` (для "безымянных" точек можно передавать по умолчанию тип ```None```) **Заданные имена точек должны быть уникальны!** (Можно реализовать на уровне проверки вхождения имени в *статическую коллекцию ```names```* реализованную на уровне класса). 

> **Выполните в классе проверку на корректность вводимых атрибутов.**

> **```Дополнительно```** Реализуйте метод класса, позволяющий построить график со всеми созданными точками и их подписанными именами.

## Задача 13.2 $\star$

**Написать собственный класс ```Line``` (```Отрезок```).**

В качестве необходимых атрибутов класс ```Line``` должен имплементировать:

* Уникальный для каждого отрезка идентификатор ```id```, определяющий порядковый номер создаваемого отрезка (Можно реализовать через инкременитруемый счетчик в *статической переменной ```count```* реализованный на уровне класса);

* Определяющие отрезок начальную ```start_point``` и конечную ```end_point``` точки класса ```Point```;

* Значения длины отрезка ```distance```  и дирекционного угла ```azimut```, которые должны рассчитываться автоматически при создании объекта, на основании заданных точек;

> **Реализуйте метод класса ```invert```, позволяющий перенаправить отрезок в обратную сторону.**

> **Реализуйте метод класса принимающий:**
* Начальную точку отрезка ```start_point``` класса ```Point```;
* Значения длины отрезка ```distance``` и дирекционного угла ```azimut```;

> **и возвращающий соответствующий введенным данным объект класса ```Line```;**

> **Реализуйте метод класса ```cross_point```, возвращающий точку пересечения (класс ```Point```) двух отрезков (для параллельных отрезков вернуть ```None```.**

> **Реализуйте метод класса ```is_crossing```, проверяющий пересекаются ли два заданных отрезка и возвращающий ```True``` если точка пересечения принадлежит обоим отрезкам и ```False``` - если нет;**

> **```Дополнительно```** Реализуйте метод класса ```fit_line```, вписывающий отрезок по МНК в набор точек класса ```Point``` и и возвращающий полученный объект класса ```Line``` (Координаты ```x``` начальной и конечной точек отрезки могут быть выбраны как $x_min$ и $x_max$ из переданных точек, а координаты ```y```, вчисленны для них по рассчитанному уравнению регрессии.

> **```Дополнительно```** Реализуйте метод класса, позволяющий построить график со всеми созданными отрезками и дополнительной возможностью вывода подписанных имен определяющих их точек.

## Задача 13.3 $\star$

**Написать собственный класс ```Polyline``` (```Полилиния```).**

По умолчанию создавать объект полилинии по переданному набору точек класса ```Point```.

В качестве необходимых атрибутов класс ```Line``` должен имплементировать:

* Уникальный для каждой полилинии идентификатор ```id```, определяющий порядковый номер создаваемой полилинии (Можно реализовать через инкременитруемый счетчик в *статической переменной ```count```* реализованный на уровне класса);

* Определяющие отрезок начальную ```start_point``` и конечную ```end_point``` точки класса ```Point```;

* Набор отрезков ```lines``` класса ```Line```, образующих полилинию.

> **Реализуйте метод класса ```distance```, позволяющий рассчитать и вернуть длину полилинии.**

> **Реализуйте метод класса, позволяющий проверить полилинию на замкнутость.**

> **Реализуйте метод класса ```invert```, позволяющий перенаправить полилинию (и все входящие в нее отрезки) в обратную сторону.**

> **```Дополнительно```** Реализуйте метод класса ```is_self_crossing```, позволяющий проверить полилинию на самопересечение.

> **```Дополнительно```** Реализуйте метод класса, позволяющий создать полилинию на основе переданного набора последовательных отрезков (Необходимо выполнить проверку на корректность последовательности отрезков - равенство координат смежных точек).

> **```Дополнительно```** Реализуйте метод класса, позволяющий построить график со всеми созданными полилиниями и дополнительной возможностью вывода подписанных имен определяющих вершины точек.
